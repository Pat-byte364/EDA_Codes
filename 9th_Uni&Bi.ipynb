```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Car Price.csv")
```

```python
plt.figure()
sns.histplot(df["Selling_Price"], kde=True)
plt.title("Price")
plt.show()
```

```python
plt.figure()
sns.boxplot(x= df["Selling_Price"])
plt.title("Expensive cars")
plt.show()
```

```python
plt.figure()
sns.countplot(x=df['Fuel'])
plt.title("Fuel Type Count")
plt.show()
```

```python
plt.figure()
plt.scatter(df['KM_Driven'], df['Selling_Price'])
plt.xlabel("KM Driven")
plt.ylabel("Selling Price")
plt.title("KM vs Price")
plt.show()
```

```python
plt.figure()
sns.boxplot(x='Fuel', y='Selling_Price', data=df)
plt.title("Fuel vs Price")
plt.show()
```
