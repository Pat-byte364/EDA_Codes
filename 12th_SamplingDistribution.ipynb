```python
import numpy as np

population = np.arange(1, 101)   # simple population (1 to 100)
```

```python
sample_size = 10
num_samples = 30

sample_means = []
```

```python
for i in range(num_samples):
    sample = np.random.choice(population, sample_size)
    mean = np.mean(sample)
    
    sample_means.append(mean)
    
    print(f"Sample {i+1} Mean = ", mean)
```

```python
print("\nAll Sample Means:")
print(sample_means)
```

```python
import matplotlib.pyplot as plt

plt.hist(sample_means, bins=10)
plt.title("Sampling Distribution of Mean")
plt.show()
```
